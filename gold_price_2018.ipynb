```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup as bs
```

```python
site=requests.get("https://www.rupeerates.in/Gold/2018")
print(site)
print(site.content)
```

```python
data=bs(site.content,"html.parser")
print(data.prettify())
```

```python
date=[]
for i in data.find_all('div',{"class":"c"})[4::4]:
    date.append(i.text)
print(date)    
```

```python
day=[]
for j in data.find_all('span',class_="day-caption"):
    day.append(j.text)
print(day)    
```

```python
price=[]
for k in data.find_all('div',class_="c text-right"):
    price.append(k.text)
print(price)    
```

```python
change=[]
for l in data.find_all('div', class_="c text-center"):
    change.append(l.text)
print(change)    
```

```python
print(len(date))
print(len(day))
print(len(price))
print(len(change))
```

```python
gold_price=pd.DataFrame({"Date":date,"Day":day,"Price Per Gram":price,"Change":change})
gold_price
```

```python
gold_price.to_csv("Gold_Price_2018.csv",index=False)
```
